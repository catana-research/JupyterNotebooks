# Advanced interval selections
https://altair-viz.github.io/user_guide/interactions.html#interval-selections

```python
# Bin data into four categories
pd.cut(df['eta'], bins=4).value_counts()
```

```python
# Find 4 bins of equal population using qcut
pd.qcut(df['pT'], q=4, precision=1).value_counts()
```

```python
df['eta_bin'] = pd.cut(df['eta'], bins=4, labels=['a', 'b', 'c', 'd']) # Giving labels to specific bins, using a uniform cut
df['pT_bin'] = pd.qcut(df['pT'], q=4)                                  # Automatically binning and labeling by quantile
df['pT_bin'] = df['pT_bin'].apply(lambda x: '({:.1f}, {:.1f}]'.format(x.left, x.right))
df.head()
```

```python
"""
The legend is a 2D grid that can be selected by shift clicking
e.g. can select Japan for a specific Cylinder, for all Cyclinders or any other selection
"""
import altair as alt

selection = alt.selection_multi(fields=['eta_bin', 'name']) # 2D legend
color = alt.condition(selection,
                      alt.Color('eta_bin:N', legend=None),
                      alt.value('lightgray'))

scatter = alt.Chart(df).mark_point().encode(
    x='pT:Q',
    y='eta:Q',
    color=color,
    tooltip='name:N'
)

scatter2 = alt.Chart(df).mark_point().encode(
    x='pT:Q',
    y='eta:Q',
    color=color,
    tooltip='name:N'
)

legend = alt.Chart(df).mark_rect().encode(
    y=alt.Y('eta_bin:N', axis=alt.Axis(orient='right')),
    x='name:O',
    color=color
).add_selection(
    selection
)

scatter | scatter2 | legend
```

# Confidence intervals

```python
import altair as alt

# draw CI
ci = alt.Chart(df).mark_area(opacity=0.3).encode(
    x='eta_bins:N',
    y='ci0(pT)',  # ci0 - Lower CI
    y2='ci1(pT)',  # ci1 - Upper CI
    color='name',
).transform_bin(
    'eta_bins', field='eta'
).properties(
    width=500
)

# Draw mean
lines = alt.Chart(df).mark_line().encode(
    x='eta_bins:N',
    y='mean(pT)',
    color='name'
).transform_bin(
    'eta_bins', field='eta', bin=alt.Bin(maxbins=10)
).properties(
    width=500
)

ci + lines
```

```python
import altair as alt
alt.data_transformers.disable_max_rows() # Disable max rows == 5000 limitation
alt.data_transformers.enable('json')     # Save data to JSON, rather than to notebook to keep size small

brush = alt.selection(type='interval', resolve='global')
#brush = alt.selection_multi(on='mouseover', toggle=False, empty='none')

base = alt.Chart(df).mark_rect(opacity=0.2).encode(
    y='eta_bins:N',
    color=alt.condition(brush, 'name', alt.ColorValue('gray')),
).add_selection(
    brush
).properties(
    width=250,
    height=250
).transform_bin(
    'eta_bins', field='eta', bin=alt.Bin(maxbins=10)
).transform_bin(
    'pT_bins', field='pT', bin=alt.Bin(maxbins=20)
).transform_bin(
    'phi_bins', field='phi', bin=alt.Bin(maxbins=20)
)

base.encode(x='pT_bins:N') | base.encode(x='phi_bins:N') | base.encode(x='charge') 
```

```python

brush = alt.selection(type='interval', resolve='global')

# Eta-Phi scatter plot
base = alt.Chart(df).mark_point(opacity=0.2).encode(
    y='eta',
    color=alt.condition(brush, 'name', alt.ColorValue('gray')),
).add_selection(
    brush
).properties(
    width=250,
    height=250
)

# Selection mu+/mu- distribution
bars = alt.Chart(df).mark_bar().encode(
    y='name:N',
    color='name:N',
    x='count(name):Q'
).transform_filter(
    brush
).properties(
    width=250
)

pt_binning = alt.Bin(step=10)

# Full pT distribution
pt_bars = alt.Chart(df).mark_bar(opacity=0.2).encode(
    x='pT_binned:N',
    y='count(pT_binned):Q'
).transform_bin(
    'pT_binned', field='pT', bin=pt_binning
).properties(
    width=250,
    height=250
)

# Selection pT distribution
pt_bars_selection = alt.Chart(df).mark_bar().encode(
    x='pT_binned:N',
    y='count(pT_binned):Q',
    color='name:N', # Can remove this to just plot total
).transform_filter(
    brush
).transform_bin(
    'pT_binned', field='pT', bin=pt_binning
).properties(
    width=250,
    height=250
)

pt_plot = alt.layer(pt_bars, pt_bars_selection)

base.encode(x='phi') & bars | pt_plot
```

```python
eta_binning = alt.Bin(step=0.5)

# Full eta distribution
eta_bars = alt.Chart(df).mark_bar(opacity=0.2).encode(
    x='eta_binned:N',
    y='count(eta_binned):Q'
).transform_bin(
    'eta_binned', field='eta', bin=eta_binning
).properties(
    width=250,
    height=250
)

# Selections
eta_lines = alt.Chart(df).mark_line().encode(
    x='eta_binned:O',
    y='count(eta_binned):Q',
    color='name:O'
).transform_bin(
    'eta_binned', field='eta', bin=eta_binning
).transform_filter(
    alt.FieldOneOfPredicate(field='name', oneOf=['mu-', 'mu+'])
)

alt.layer(eta_bars, eta_lines)
```

```python
"""
Perform, non-rectangular selections by transforming the data
"""
brush = alt.selection(type='interval', resolve='global')

# Eta-Phi trasnformed polar coordinate scatter plot
scatter_transform = alt.Chart(df).mark_point(opacity=0.2).encode(
    x='theta:Q',
    y='r:Q',
    color=alt.condition(brush, 'name', alt.ColorValue('gray')),
).properties(
    width=250,
    height=250
).transform_calculate(
    theta='atan(datum.eta/datum.phi)',
    r='sqrt(datum.eta*datum.eta + datum.phi*datum.phi)'
).add_selection(
    brush
)

# Eta-Phi polar coordinate scatter plot
scatter = alt.Chart(df).mark_point(opacity=0.2).encode(
    x='eta:Q',
    y='phi:Q',
    color=alt.condition(brush, 'name', alt.ColorValue('gray')),
).properties(
    width=250,
    height=250
).transform_calculate(
    theta='atan(datum.eta/datum.phi)',
    r='sqrt(datum.eta*datum.eta + datum.phi*datum.phi)'
).transform_filter(
    brush
)


scatter_transform | scatter
```

```python
"""
Bind table to selection
"""

import altair as alt
from vega_datasets import data

source = data.cars()

# Brush for selection
brush = alt.selection(type='interval')

# Scatter Plot
points = alt.Chart(df).mark_point().encode(
    x='eta:Q',
    y='phi:Q',
    color=alt.condition(brush, 'pT:Q', alt.value('grey'))
).add_selection(brush)

# Base chart for data tables
ranked_text = alt.Chart(df).mark_text().encode(
    y=alt.Y('row_number:O',axis=None)
).transform_window(
    row_number='row_number()'
).transform_filter(
    brush
).transform_window(
    rank='rank(row_number)'
).transform_filter(
    alt.datum.rank<20
)

# Data Tables
eta = ranked_text.encode(text='eta:Q').properties(title='Eta')
phi = ranked_text.encode(text='phi:Q').properties(title='Phi')
pT = ranked_text.encode(text='pT:Q').properties(title='pT')
text = alt.hconcat(eta, phi, pT) # Combine data tables

# Build chart
alt.hconcat(
    points,
    text
).resolve_legend(
    color="independent"
)
```

```python
import altair as alt

slider = alt.binding_range(min=0, max=200, step=1, name='cutoff:')
selector = alt.selection_single(name="SelectorName", fields=['cutoff'],
                                bind=slider, init={'cutoff': 50})

# Only colour muons blue with pT above cutoff, otherwise red
scatter1 = alt.Chart(df).mark_point().encode(
    x='eta',
    y='phi',
    color=alt.condition(
        alt.datum.pT < selector.cutoff,
        alt.value('red'), alt.value('blue')
    )
).add_selection(
    selector
)


# Only show muons with pT above cutoff
scatter2 = alt.Chart(df).mark_point().encode(
    x='eta',
    y='phi',
    color='pT'
).add_selection(
    selector
).transform_filter(
        alt.datum.pT > selector.cutoff,
)

scatter1 | scatter2
```

```python
"""
Histogram selections
"""

import altair as alt

brush = alt.selection(type='interval', encodings=['x'])

# Define the base chart, with the common parts of the
# background and highlights
base = alt.Chart().mark_bar().encode(
    x=alt.X(alt.repeat('column'), type='quantitative', bin=alt.Bin(maxbins=20)),
    y='count()'
).properties(
    width=160,
    height=130
)

# blue background with selection
background = base.add_selection(brush)

# yellow highlights on the transformed data
highlight = base.encode(
    color=alt.value('goldenrod')
).transform_filter(brush)

# layer the two charts & repeat
alt.layer(
    background,
    highlight,
    data=df
).repeat(column=["eta", "phi", "pT"])
```
