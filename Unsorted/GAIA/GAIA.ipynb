# Data Model

Full documentation: https://gaia.esac.esa.int/documentation/GDR1/pdf/GaiaDR1_datamodel_1.2.pdf


Summary:

- <b>source id</b> : Unique source identifier <i>(long)</i>
A unique single numerical identifier of the source
- ref epoch : Reference epoch <i>(double, Time[Julian Years])</i>
Reference epoch to which the astrometic source parameters are referred, expressed
as a Julian Year in TCB.
- <b>ra</b> : Right ascension <i>(double, Angle[deg])</i>
Barycentric right ascension α of the source in ICRS at the reference epoch ref epoch
- <b>ra</b> error : Standard error of right ascension <i>(double, Angle[mas])</i>
Standard error σα∗ ≡ σα cos δ of the right ascension of the source in ICRS at the
reference epoch ref epoch.
- <b>dec</b> : Declination <i>(double, Angle[deg])</i>
Barycentric declination δ of the source in ICRS at the reference epoch ref epoch
- <b>dec</b> error : Standard error of declination <i>(double, Angle[mas])</i>
Standard error σ$_{δ}$ of the declination of the source in ICRS at the reference epoch
ref epoch
- <b>parallax</b> : Parallax <i>(double, Angle[mas] )</i>
Absolute barycentric stellar parallax $\overline{\omega}$ of the source at the reference epoch ref epoch
- <b>parallax error</b> : Standard error of parallax <i>(double, Angle[mas] )</i>
Standard error σ$_{\overline{\omega}}$ of the stellar parallax at the reference epoch ref epoch
- <b>phot g mean flux</b> : G-band mean flux <i>(double, Flux[e-/s])</i>
Mean flux in the G-band.
- <b>phot g mean flux error</b> : Error on G-band mean flux <i>(double, Flux[e-/s])</i>
Error on the mean flux in the G-band.
- <b>phot g mean mag</b> : G-band mean magnitude <i>(double, Magnitude[mag])</i>
Mean magnitude in the G band. This is computed from the G-band mean flux applying
the magnitude zero-point in the Vega scale
- <b>l</b> : Galactic longitude <i>(double, Angle[deg])</i>
Galactic Longitude of the object at reference epoch ref epoch, see ESA, 1997,
’The Hipparcos and Tycho Catalogues’, ESA SP-1200, Volume 1, Section 1.5.3, for
the conversion details.
- <b>b</b> : Galactic latitude <i>(double, Angle[deg])</i>
Galactic Latitude of the object at reference epoch ref epoch, see ESA, 1997, ’The
Hipparcos and Tycho Catalogues’, ESA SP-1200, Volume 1, Section 1.5.3, for the conversion
details.
- <b>ecl lon</b> : Ecliptic longitude <i>(double, Angle[deg])</i>
Ecliptic Longitude of the object at reference epoch ref epoch, see ESA, 1997,
’The Hipparcos and Tycho Catalogues’, ESA SP-1200, Volume 1, Section 1.5.3, for
the conversion details.
- <b>ecl lat</b> : Ecliptic latitude <i>(double, Angle[deg])</i>
Ecliptic Latitude of the object at reference epoch ref epoch, see ESA, 1997, ’The
Hipparcos and Tycho Catalogues’, ESA SP-1200, Volume 1, Section 1.5.3, for the conversion
details.

## Suplementary information

- In 1° there are 60 arcminutes, written 60'. One arcminute contains 60 arcseconds, written 60".

- Paralax: 
$$d({\rm pc}) = \frac{1}{p({\rm arcsec})}$$

- An object's absolute magnitude is defined to be equal to the apparent magnitude that the object would have if it were viewed from a distance of exactly 10 parsecs.

- Lowest apparent magnitude visible by human eye is approximately 6.5, values above this are not visible.
